```python
import pandas as pd
import json
```

```python
df = pd.read_csv("../data/kyiv-channels-dataset/KyivChannels_Dataset_v01_lang.csv")
```

```python
# among duplicated contents remain only one in each group
df = df[~df['content'].duplicated()]
```

```python
channels = df['channelname'].unique()
```

```python
def preprocess(string):
    string = " ".join(string.split())
    string = string.strip()
    return string 
```

```python
stoprows = {}
FREQ_TH = 10 
skip_channels = ["uhmc2022", "kyivpasstrans", "kievvlast", "kyivcityofficial"]

for channel in channels:
    if channel in skip_channels:
        continue

    df_channel = df[df['channelname'] == channel]
    rows = []

    for text in df_channel['content']:
        rows.extend(text.split("\n"))

    rows = pd.Series(rows).apply(preprocess)
    rows_counts = rows.value_counts()
    stop_rows_channel = list(set(rows_counts[rows_counts >= FREQ_TH].index))
    stoprows[channel] = stop_rows_channel
```

```python
json.dump(stoprows, open("../models/stoprows.json", 'w'), ensure_ascii=False)
```
